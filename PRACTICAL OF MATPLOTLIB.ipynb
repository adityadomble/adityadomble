```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Select the numerical columns of interest
numerical_cols = ['age', 'fare']

# Iterate through each numerical column
for col in numerical_cols:
    print(f"Statistics for '{col}':")

    # Minimum
    min_val = titanic[col].min()
    print(f"  Minimum: {min_val}")

    # Maximum
    max_val = titanic[col].max()
    print(f"  Maximum: {max_val}")

    # Quartiles
    Q1 = titanic[col].quantile(0.25)  # First quartile (25th percentile)
    Q2 = titanic[col].quantile(0.50)  # Second quartile (50th percentile), also the median
    Q3 = titanic[col].quantile(0.75)  # Third quartile (75th percentile)

    print(f"  Q1 (25th percentile): {Q1}")
    print(f"  Q2 (Median, 50th percentile): {Q2}")
    print(f"  Q3 (75th percentile): {Q3}")

    # IQR
    IQR = Q3 - Q1
    print(f"  IQR: {IQR}")

    #Upper and Lower Bounds
    upper_bound = Q3 + (1.5 * IQR)
    lower_bound = Q1 - (1.5 * IQR)
    print(f"  Upper Bound: {upper_bound}")
    print(f"  Lower Bound: {lower_bound}")

    #Outliers
    outliers = titanic[(titanic[col] < lower_bound) | (titanic[col] > upper_bound)]
    print(f"  Number of Outliers: {len(outliers)}")

    print("-" * 30)  # Separator between columns
```
